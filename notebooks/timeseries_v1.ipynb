```python
import altair as alt
import pandas as pd
import statistics
dat = pd.read_csv("./rat_dat.csv")
print(len(dat))
# alt.data_transformers.enable('csv')

def compress(a, f, step):
    ret = []
    accum = 0
    for i in range(0,len(a),step) :
        if i == 0:
            continue
        t = a[accum:i]
        ret.append(f(t))
        accum+=step
    return ret

vals = compress(dat.values[:,1].tolist(), statistics.mean, 1000)
compressed_data = [list(a) for a in zip(range(len(vals)), vals)]
df = pd.DataFrame(compressed_data, columns = ['Time', 'Data'])

print(len(df))
chart = alt.Chart(df).mark_line().encode(
    x='Time',
    y='Data').interactive()

chart.configure_view(
    continuousHeight=600,
    continuousWidth=2000,
)
```
